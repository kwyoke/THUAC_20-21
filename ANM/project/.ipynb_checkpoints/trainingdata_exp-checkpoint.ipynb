# Import and setup

```python
import os, sys
import glob
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
trace_files = glob.glob("training_data/2020_05_04/trace/*.csv")
print(trace_files)
host_files = glob.glob("training_data/2020_05_04/host/*.csv")
print(host_files)
esb_file = "training_data/2020_05_04/esb.csv"
```

```python
def get_df_fr_csv(all_files):
    li = []

    for filename in all_files:
        df = pd.read_csv(filename, index_col=None, header=0)
        li.append(df)

    df_all = pd.concat(li, axis=0, ignore_index=True)
    return df_all
```

## Fishy... itemid != name_cmdb

I thought each unique itemid corresponds to each unique {name, cmdb_id} but seems like it's not the case. In general, each unique itemid only matches to one name_cmdb, but different itemids could match to the same name_cmdb. {name, cmdb_id} containing Disk_io_util, Incoming_network_traffic, Outgoing_network_traffic, New_Tbs_Free_Gb (non-exhaustive) seem to be associated with multiple itemids. So what exactly is itemid supposed to mean?

```python
uniq_items = df_host['itemid'].unique()
same_namecmdb = True
for item in uniq_items:
    db_item = df_host[df_host['itemid'] == item]
    if len(db_item['name_cmdb'].unique()) > 1:
        print(item, len(db_item['name_cmdb'].unique())
        same_namecmdb = False
if same_namecmdb:
    print('Each item maps to one name_cmdb')
```

```python
uniq_namecmdb = df_host['name_cmdb'].unique()
same_item = True

for i in uniq_namecmdb:
    db_namecmdb = df_host[df_host['name_cmdb'] == i]
    if len(db_namecmdb['itemid'].unique()) > 1:
        print(i, len(db_namecmdb['itemid'].unique()))
        same_item = False

if same_item:
    print('Each name_cmdb maps to one item')
else:
    print('Each name_cmdb maps to more than one item')
```

## KDE for KPI_Host in short time window

```python
df_host = get_df_fr_csv(host_files)
df_host['name_cmdb'] = df_host['name'] + '+' + df_host['cmdb_id']

items = ['os', 'redis', 'db', 'container', 'docker']
host_type = []
for el in df_host['cmdb_id']:
    for search in items:
        if el.startswith(search):
            host_type.append(search)
df_host['host_type'] = host_type

df_host.head()
```

```python
sec = 1000 #ms
minute = 60*sec
hr = 60*minute
start = 1588521600000
end = 1588607999000
df_win1h = df_host[(df_host['timestamp'] > start+3*hr) & (df_host['timestamp'] <= start+4*hr)]
df_win10min = df_host[(df_host['timestamp'] > start+3*hr+50*minute) & (df_host['timestamp'] <= start+4*hr)]
```

```python
from sklearn.neighbors import KernelDensity
```

```python
def vis_kde(name_cmdb, df_win1h, df_win10min, bw = 1.0):
    # KDE and fitting on 1h window
    kpi_host_samp = df_win1h[df_win1h['name_cmdb']==name_cmdb][['timestamp', 'value']]
    x = kpi_host_samp['value']

    # instantiate and fit the KDE model
    kde = KernelDensity(bandwidth=bw, kernel='gaussian')
    kde.fit(x[:, None])

    # Find -logprob for 10min window
    kpi_host_samp10min = df_win10min[df_win10min['name_cmdb']==name_cmdb][['timestamp', 'value']]
    x10min = kpi_host_samp10min['value']
    # score_samples returns the log of the probability density
    logprob = -1*kde.score_samples(x10min[:, None])
    kpi_host_samp10min['logprob'] = logprob

    plt.subplot(2,1,1)
    '''
    plt.title(name_cmdb + ' before KDE')
    sns.lineplot(x="timestamp", y="value", data=kpi_host_samp, alpha=0.5)
    plt.ylabel('Value')
    plt.xlabel('Timesteps')
    '''
    
    plt.plot(x, np.full_like(x, -0.01), '|k', markeredgewidth=1)

    plt.subplot(2,1,2)
    plt.title(name_cmdb + ' after KDE')
    sns.lineplot(x="timestamp", y="logprob", data=kpi_host_samp10min, alpha=0.5)
    plt.ylabel('Negative of logprob')
    plt.xlabel('Time')
    plt.tight_layout()
    return kpi_host_samp10min
```

```python
logprob = vis_kde('New_Tbs_Used_Pct+db_006', df_win1h, df_win10min)
```

### Time-series KPI graphs

Generally not much seasonality for KPI values. Some got a bit, but its generally a flat value across 24hr. Occasional deviations are probably anomalies. 

KPI value distribution very different across different hosts. So if we want to model the distribution we need to maintain 2k KDE estimations...

```python
def visualise_KPI(start=0, end=16, i_16=None, single_index=None):
    if i_16 is not None:
        start = i_16 * 16
        end = (i_16+1) * 16
    
    if single_index is not None:
        start = single_index
        end = single_index + 1
        
    kpi_names = df_host['name'].unique()
    i=0
    print("index", start, "to", end)
    for kpi in kpi_names[start: end]:
        i+=1
        pf_1 = df_host[df_host.name == kpi]
        if end-start>1:
            ax=plt.subplot(4,4,i)
            legend = False
        else:
            ax=plt.subplot(1,1,i)
            legend = True
            plt.title(kpi)
        sns.lineplot(x="timestamp", y="value", hue="cmdb_id", data=pf_1, ax=ax, legend=legend, alpha=0.5)
```

```python
def checkIndex(name):
    return np.where(df_host['name'].unique() == name)[0][0]
```
