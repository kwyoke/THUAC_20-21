```python
import requests
from bs4 import BeautifulSoup
import random
import re
import json
```

## SERP INFO FILE

```python
def get_bing_top10(query_text):
    top10 = []
    prev_info = []
    url_base = 'https://www.bing.com/search?q=' 
    A = ("Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36",
           "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2227.1 Safari/537.36",
           "Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2227.0 Safari/537.36",
           )

    Agent = A[random.randrange(len(A))]
    headers = {'user-agent': Agent}
    
    while len(top10) < 10:
        first = len(top10)
        qfirst = '&first=' + str(first)
        url = url_base + query_text + qfirst #to adjust number of queries shown
        
        r = requests.get(url, headers=headers)
        soup = BeautifulSoup(r.text, 'lxml')

        rel_results = soup.find_all('li', class_='b_algo')
        i = 0
        for info in rel_results:
            if i==0:
                if info in prev_info:
                    continue
            i+=1
            rank = i+first
            title = info.find('a').text
            link = info.find('a')['href']
            res_dict = {'rank': rank, 
                        'title': title, 
                        'url': link}
            top10.append(res_dict)
            if len(top10) ==10:
                break
        prev_info = rel_results
    return top10
```

```python
def get_google_top10(query_text):
    top10 = []
    prev_info = []
    url_base = 'https://www.google.com/search?q=' 
    A = ("Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36",
           "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2227.1 Safari/537.36",
           "Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2227.0 Safari/537.36",
           )

    Agent = A[random.randrange(len(A))]
    headers = {'user-agent': Agent}
    
    first = len(top10)
    qfirst = '&start=' + str(first)
    url = url_base + query_text + qfirst #to adjust number of queries shown

    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, 'lxml')

    rel_results = soup.find_all('div', class_="ZINbbc xpd O9g5cc uUPGi")
    i = 0
    for info in rel_results:
        if i==0:
            if info in prev_info:
                continue
        
        title = info.find('h3', class_="zBAuLc")
        link = info.find('div', class_="kCrYT")
        if None in (title, link):
            continue
        
        i+=1
        rank = i+first
        title = title.text
        link = link.find('a')['href']
        m = re.search('http(.+?)&sa', link) #use regex to get url in correct format, quite ugly here
        if m:
            tmp = m.group(1)
            link = "http"
            link += tmp
        res_dict = {'rank': rank, 
                    'title': title, 
                    'url': link}
        top10.append(res_dict)
        if len(top10) ==10:
            break
    prev_info = rel_results
    return top10
```

```python
def write_serp_info(serpdict, outfilename):
    with open(outfilename, 'w') as outfile:
        json.dump(serpdict, outfile)
```

# Crawl target pages

```python
def save_page(url, outfilename):
    A = ("Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36",
           "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2227.1 Safari/537.36",
           "Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2227.0 Safari/537.36",
           )

    Agent = A[random.randrange(len(A))]
    headers = {'user-agent': Agent}
    r = requests.get(url, headers=headers)
    if 'html' in r.headers.get('content-type'):
        open(outfilename + '.html', 'wb').write(r.content)
        return
    if 'pdf' in r.headers.get('content-type'):
        open(outfilename + '.pdf', 'wb').write(r.content)
        return
    print('URL is neither html nor pdf')
```

```python
def crawl_target_results(serpdict, se='BING',querynum='1'):
    for res in serpdict:
        outfilename = 'Target_Pages_1_2/TP_' + se + '_' + querynum + '_' + str(res['rank']) + '_2020280598'
        save_page(res['url'], outfilename)
```

# MAIN

```python
q1 = 'python request ssl certificate error'
q2 = 'bitcoins digital currency gamestop saga'

bing_q1 = get_bing_top10(q1)
bing_q2 = get_bing_top10(q2)
google_q1 = get_google_top10(q1)
google_q2 = get_google_top10(q2)

outfilenames = ['json/SE_BING_1_2020280598.json', 'json/SE_BING_2_2020280598.json', 'json/SE_GOOGLE_1_2020280598.json','json/SE_GOOGLE_2_2020280598.json']
write_serp_info(bing_q1, outfilenames[0])
write_serp_info(bing_q2, outfilenames[1])
write_serp_info(google_q1, outfilenames[2])
write_serp_info(google_q2, outfilenames[3])
```
